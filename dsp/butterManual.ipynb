```python
import numpy as np
from math import sqrt,pow,log10,pi
import matplotlib.pyplot as plt

def butt(order=1,cutFreq=1):
    freqs=np.linspace(0.01,100,1000)
    butter=[]
    for i in freqs:
        butter.append(1/(sqrt(1+(pow((i/cutFreq),(2*order))))))

    orderLabel=f"Ordem:{order}"

    plt.grid()
    plt.plot(np.log10(freqs),20*np.log10(butter),label=orderLabel)
    plt.xlabel('Log da Frequência')
    plt.ylabel('Ganho (dB)')
```

```python

cutFreq=1

for i in range(1,6):
    butt(order=i,cutFreq=cutFreq)

plt.axvline(np.log10(cutFreq), color='green',label=f"F de corte:{cutFreq}")
plt.ylim(bottom=-100)
plt.xlim(right=2,left=-2)
plt.legend()
plt.show()
```
